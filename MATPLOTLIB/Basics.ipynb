```python
x=[1,2,3,4,5,6,7,8]
y=[10,40,20,50,60,20,90,65]

plt.plot(x,y)

plt.show()
```

## PYPLOT
plt.plot(x,y, color="", linestyle="", linewidth="", marker="marker-symbol", label="")

```python
x=["mon","tue","wed","thus","fri"]
y=[20,15,25,30,10]

plt.plot(x,y)
plt.title("Bakery Sales Data")
plt.xlabel("days of the week")
plt.ylabel("sales per day")
```

```python
months=["jan","feb","march","april","may"]
sales=[2500,3000,2000,1500,2700]
plt.plot(months,sales,color="blue", linestyle="--",linewidth=2,marker="o",label="2025 sales data")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("2025 sales data")
plt.legend(loc="upper right" , fontsize=10)
plt.grid(color="grey" , linestyle=":" , linewidth=1)
plt.xlim("jan","may")
plt.ylim(1000,3000)
plt.show()
```

# BAR GRAPH
plt.bar(x,y,color="")

```python
months=["jan","feb","march","april","may"]
sales=[2500,3000,2000,1500,2700]
plt.bar(months,sales,color="orange")
plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.title("Monthy Sales")
plt.legend()
plt.show()
```

# Horizontal Bar graph

```python
months=["jan","feb","march","april","may"]
sales=[2500,3000,2000,1500,2700]
plt.barh(months,sales,color="orange")
plt.xlabel("Months")
plt.ylabel("Sales per month")
plt.title("Monthy Sales")
plt.legend()
plt.show()
```

# Pie Chart
plt.pie(value,label="partition",autopct="%1.1%f%%",colors=[])

```python
regions=["north","south","east","west"]
revenue=[2000,4000,1500,3000]
plt.pie(revenue,labels=regions,autopct="%1.1f%%", colors=["blue","green","yellow","red"])
plt.title("revenue contribution by region")
plt.show()
```

# Histogram
plt.hist(value,bins="",color="",edgecolor="")

```python
marks=[45,65,76,46,35,87,67,57,53,51,55,74,47,86,58,41,43]
plt.hist(marks,bins=5,color="purple" , edgecolor="black")
plt.xlabel("marks of students")
plt.ylabel("no of students")
plt.title("marks of  all students")
plt.show()
```

# Scatter plot
plt.scatter(x,y,color="",marker="marker style")

```python
study_hours=[1,2,3,4,5,6,7,8]
scores=[60,63,70,72,75,78,85,90]
plt.scatter(study_hours,scores,color="red",marker="o")
plt.xlabel("Study_Hours")
plt.ylabel("Marks in exam")
plt.title("Relationship B/w Study_time and Marks")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.scatter([1,2,3],[50,58,65],color="red",marker="o",label="Section A")
plt.scatter([1,2,3],[45,54,49],color="blue",marker="o",label="Section B")
plt.xlabel("Study_Hours")
plt.ylabel("Marks in exam")
plt.title("Relationship B/w Study_time and Marks")
plt.grid(True)
plt.legend()
plt.show()
```

# Subplot
plt.subplot(nrows,ncols,index)

```python
x=[1,2,3,4]
y=[10,20,15,25]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("line chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("bar chart")
plt.show()
```

fig, ax = plt.subplot(nrows , ncols , figsize=(width,height))

```python
fig , ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x,y, linestyle="--",color="red")
ax[0].set_title("line chart")

ax[1].bar(x,y,color="red")
ax[1].set_title("bar chart")
plt.suptitle("comparison of line and bar chart")
plt.tight_layout()
plt.show()
```

# Saving the graph

```python
fig , ax = plt.subplots(1,2,figsize=(10,5))
ax[0].plot(x,y, linestyle="--",color="red")
ax[0].set_title("line chart")

ax[1].bar(x,y,color="red")
ax[1].set_title("bar chart")
plt.suptitle("comparison of line and bar chart")
plt.tight_layout()

plt.savefig("line & bar graph.png",dpi=300,bbox_inches="tight")
plt.show()
```
